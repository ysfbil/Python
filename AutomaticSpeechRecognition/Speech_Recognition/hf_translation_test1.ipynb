```python
import re

cleannr=re.compile('<.*?>')

def cleanhtml(raw_text):
    cleantext=re.sub(cleannr,'',raw_text)
    return cleantext
```

```python
#asr = pipeline("automatic-speech-recognition", model="ceyda/wav2vec2-base-760-turkish")
#asr = pipeline("automatic-speech-recognition", model="m3hrdadfi/wav2vec2-large-xlsr-turkish")
asr = pipeline("automatic-speech-recognition", model="mpoyraz/wav2vec2-xls-r-300m-cv6-turkish")
```

```python
audio_file="tr_ses1.flac"
translation = asr(audio_file)

print(translation)
```

```python
print(cleanhtml(translation["text"]))
```
